```python
def getreviewids(domain, activitytype, cityid, activityid):
    baseurl = 'http://www.tripadvisor.com' + domain + '/'+activitytype+'-g'
    orstep = 10
    reviewids = set()
    yorumlar = []
    activitypage = -1
    tur = True
    dongu = 15
    while tur:
        activitypage += 1 # her on sayfada ilerle (yorumlar her sayfada 10 olacak şekilde bulunmaktadır)
        print(f'page: {activitypage} --- döngü: {dongu}')
        if activitypage >= dongu:
            break
        if activitypage == 0:
            activitiyurl = '%s%s-d%s' % (baseurl, cityid, activityid)
        else:
            activitiyurl = '%s%s-d%s-or%s' % (baseurl, cityid, activityid, activitypage * orstep)
            
        print(activitiyurl)  
        htmlpage = s.get(activitiyurl, allow_redirects=False)
        if htmlpage.text == '':
            htmlpage = s.get(activitiyurl, allow_redirects=True)
        print(f"{activitypage}. tamam")
        print("html response: ",htmlpage.status_code)
        sleep(random.randint(10,20)/10) #her işlemden sonra biraz uyumalısın
        if htmlpage.status_code != 200:
            continue
        else:
            sites = BeautifulSoup(htmlpage.text, 'html.parser')
            bul = sites.find_all('script', {'type': 'application/ld+json'})
            jsText = re.findall(r'{.*]}', str(bul[0]))
            print(f"jsText boyutu: {len(jsText)}")
            if jsText == []:
                break
            else:
                govde = json.loads(jsText[0])['review'] # yorum gövdesi json olarak
                yorumlar.append(govde)
                sayi = sites.find_all('div', attrs={'class' : 'pageNumbers'})
                dongu = len(re.findall(r'a class', str(sayi)))
            
    print(f"yorumlar boyutu: {len(yorumlar)}")
             
    return yorumlar  # , reviewids
```

```python
secondHome = pd.read_csv("./Palma.csv")
#secondHome.sort_values(by="seconHomes", inplace=True)
```

```python
for item in secondHome.seconHomes:
    ids = re.findall(r'[0-9]+', item)
    govde = getreviewids('', "VacationRentalReview", str(ids[0]), str(ids[1]))
    if len(govde) > 0:
        df = pd.concat([pd.DataFrame.from_dict(d) for d in govde], ignore_index=True)
        df['bolge'] = ids[0]
        df['house'] = ids[1]
        df.to_csv("./PalmaYorumlar.csv", index=False, mode="a")
```
