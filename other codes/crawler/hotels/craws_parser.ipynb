```python
import urllib.request as request
import os
import glob
import codecs
import csv
from typing import List, Any
from bs4 import BeautifulSoup
from time import sleep
from contextlib import closing
import re
import json
import pandas as pd
import numpy as np
```

```python
fieldname = ['bolgeID', 'otelID', 'yorumIDs']
with codecs.open("giritYorumlariTopla.csv", encoding="utf8", mode="r") as csvfile:
    idler = csv.DictReader(csvfile, fieldnames=fieldname )
    for _ in range(22480):
        next(idler)
    yorumlar = []
    for ay in idler:
        yorumlar.append([ay['bolgeID'], ay['otelID'], ay['yorumIDs']])

for k in range(len(yorumlar)):
    url = "https://www.tripadvisor.com/ShowUserReviews-g" +str(yorumlar[k][0])+"-d"+str(yorumlar[k][1])+"-r"+str(yorumlar[k][2])
    #if k==2:
    #    break
    try:
        
        with closing(request.urlopen(url)) as f:
            htmlpage = f.read().decode('utf-8')
            sleep(0.1) 
            sites = BeautifulSoup(htmlpage, 'html.parser')
            bul = sites.find_all('script', {'type': 'application/ld+json'})
            if bul==[]:
                continue
            for container in sites.find_all('div', attrs={'id': 'review_'+str(yorumlar[k][2])}):
                memleket = container.find('div', {'class': 'userLoc'}) #kullanıcı memleketi
                trip_type = container.find('div', {'class': 'recommend-titleInline'}) #seyahat türü
                if len(container.find_all('p', {'class': 'partial_entry'}))<2:
                    responseText = ""
                else:
                    responseText = container.find_all('p', {'class': 'partial_entry'})[1].text     
                responseDate = container.find('span', attrs={'class': 'responseDate'})
                stayDate = container.find('div', {'class': 'prw_reviews_stay_date_hsx'}) # otelde kaldığı tarih
                    
            jsText = re.findall(r'{.*?}}}}|{.*?}}}', str(bul))[0]                
            yorum = json.loads(jsText) # yorum gövdesi json olarak
            yorum_baslik = yorum['name']
            metin = yorum['reviewBody'] #yorum metni
            rating = yorum['reviewRating']['ratingValue']
                
            if 'datePublished' in yorum.keys():
                tarihi = yorum['datePublished']
            else:
                tarihi = sites.find('span', attrs={'class': 'ratingDate'}).attrs['title']

            yorumYaz =  List['Any']                
            yorumYaz = [yorumlar[k][1], yorumlar[k][2], '' if memleket==None else memleket.text,
                        tarihi, str(yorum_baslik), str(metin), 
                        '' if stayDate==None else stayDate.text, rating,
                        '' if trip_type==None else trip_type.text,
                        responseDate.text if responseDate!=None else '' , 
                        responseText
                       ]
            pd.DataFrame(np.reshape(yorumYaz, (1,11))).to_csv("c:/data/giritSon.csv", mode="a", index=False, header=False)
            print(k, '. yorum')
    except:
        print("url hatası")
```
