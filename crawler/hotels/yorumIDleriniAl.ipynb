```python
import csv
import os
import glob

def IDAl(path):
    with open(path, newline="\n", encoding="ISO-8859-9") as csvfile:
        line = csv.reader(csvfile, delimiter=",")
        seen = set()
        oteller = []
        for al in line:
            if al[1] not in seen:
                seen.add(al[1])
                oteller.append([al[0], al[1]])
    return oteller
path="D:/data/yeniAralik2019/en/"
allFiles = glob.glob(os.path.join(path, "*.csv"))
allIDs = []
for file_ in allFiles:
    allIDs += IDAl(file_)
```

```python
#np.savetxt("d:/aa.txt", pd.DataFrame(allIDs))
pd.DataFrame(allIDs).drop_duplicates().to_csv("d:/data/yeniAralik2019/otelIds.csv", index=False, header=False)
```

```python
import os
import re
import codecs
import csv
from typing import List, Any

import urllib.request as request
import urllib.error as urlerror
import socket
from contextlib import closing
from time import sleep


def download_page(url):
    tries = 0
    htmlpage = None
    while htmlpage is None:
        try:
            with closing(request.urlopen(url)) as f:
                htmlpage = f.read()
                sleep(2)
        except (urlerror.URLError, socket.timeout, socket.error):
            tries += 1
    return htmlpage


def getreviewids(domain, activitytype, cityid, activityid):
    baseurl = 'http://www.tripadvisor.com' + domain + '/'+activitytype+'_Review-g'
    orstep = 5
    reviewids = set()
    activitypage = 0
    reviewre = re.compile(r'/ShowUserReviews-g%s-d%s-r([0-9]+)-' % (cityid, activityid))

    while True:
        if activitypage == 0:
            activitiyurl = '%s%s-d%s' % (baseurl, cityid, activityid)
        else:
            activitiyurl = '%s%s-d%s-or%s' % (baseurl, cityid, activityid, activitypage * orstep)

        htmlpage = download_page(activitiyurl)

        if htmlpage is None:
            print('Error downloading the URL: ' + activitiyurl)
            break
        else:
            pageids = set(reviewre.findall(htmlpage.decode('utf-8')))
            allin = True
            for id in pageids:
                if id not in reviewids:
                    allin = False
                    break

            if allin:
                break
            pageids = list(pageids)
            pageids = set(pageids)

            reviewids.update(pageids)
            activitypage += 1

    return reviewids




def main():
    idpath = "c:/data/yeniAralik2019/trIDs.csv"
    fs = open(idpath, newline="\n")
    line = csv.reader(fs, delimiter=",")
    i=0
    for ids in line:
        reviews = [ids[0], ids[1], getreviewids(".tr", "Hotel", ids[0], ids[1])]
        #with open("D:/data/yeniAralik2019/ilk600.csv", mode="a") as ilk600:
        #   yazici = csv.writer(ilk600, delimiter="\n")
        #   yazici.writerow(reviews)
        pd.DataFrame(np.reshape(reviews, (1,3))).to_csv("c:/data/yeniAralik2019/trKalan.csv", mode="a", index=False, header=False)
        i+=1
        print("{}. otel tamam".format(i))
    
    return 0
if __name__ == '__main__':
    main()

```
