```python
def yorumAl(url):
    yorumlar = []
    r = s.get(url, allow_redirects=True)
    sites = BeautifulSoup(r.text, 'html.parser')
    container = sites.find_all('div', attrs={'id': re.compile(r'review_[0-9]+')})
    orebek = sites.find('span', attrs={'class': 'reviews_header_count block_title'}).text[1:-1]
    for item in container:
        yorum_id = re.search(r'[0-9]+', str(item.attrs['id'])).group()
        tarih = item.find('span', attrs={'class': re.compile(r'ratingDate')}).text.strip()
        yaz = [yorum_id, tarih]
        yorum_id_date.append(yaz)
    return yorumlar, orebek
```

```python
yorum_id_date = []
for v in Fethiye:
    orstep = 10
    dongu = 0
    orebek = None
    bak = orstep*dongu
    while True:
        
        if bak == 0:
            url = "https://www.tripadvisor.com/VacationRentalReview-g298031-d" + str(v)
            al, orebek= yorumAl(url)
            yorum_id_date.append(al)
            dongu+=1
        bak = orstep*dongu
        if bak >= int(orebek):
            break        
        else:
            url = "https://www.tripadvisor.com/VacationRentalReview-g298031-d" + str(v) + "-or" + str(bak)
            al, orebek= yorumAl(url)
            yorum_id_date.append(al)
            dongu+=1
```

```python
son = [y for y in yorum_id_date if y != []]
len(son)
```

```python
col_names = ['yorumIDs', 'bolge', 'house', 'name', 'tarih', 'reviewBody', 'rating']
for i in range(df.shape[0]):
    url = "https://www.tripadvisor.com/ShowUserReviews-g" +str(df.iloc[i, :][1])+"-d"+str(df.iloc[i, :][2])+"-r"+str(df.iloc[i, :][0])
    r = s.get(url, allow_redirects=True)
    sites = BeautifulSoup(r.text, 'html.parser')
    ids = list(set(re.findall(r'review_([0-9]+)', r.text))) # her sayfada 7 tane yorum id si var.
    bak = False
    k = 0
    for al in ids:
        k+=1
        if bak:
            yorumIDleri = pd.read_csv("./secondEv_yorumlar/Palma_yorumlar.csv", header=None, names=col_names)
            yorumIDleri.drop_duplicates(subset='yorumIDs', inplace=True, ignore_index=True)
            if al in list(yorumIDleri.yorumIDs):
                continue
        container = sites.find('div', attrs={'id': 'review_' + al})
        if len(container.find('div', {'class': "entry vrReviewText"}).text) < 3:
            continue
        else:
            reviewBody = container.find('div', {'class': 'entry vrReviewText'}).text.strip()
            yorum_baslik = container.find('div', {'class': "quote vrClickTestQuote"}).text
            tarihi = container.find('span', attrs={'class': 'ratingDate'}).text
            rating = container.find('span', attrs={'class': re.compile(r"ui_bubble_rating bubble_([0-9]+)")}).attrs['class'][1].split('_')[1]
            yorumYaz =  List['Any']                
            yorumYaz = [al, str(df.iloc[i, :][1]), str(df.iloc[i, :][2]),
                        yorum_baslik, tarihi, reviewBody, rating ]
            pd.DataFrame(np.reshape(yorumYaz, (1,7))).to_csv("./secondEv_yorumlar/Palma_yorumlar.csv", mode="a", index=False, header=False)
            bak = True
        print(f"{k} yorum eklendi")
    print(f"{i+1}. ev tamam")
```
